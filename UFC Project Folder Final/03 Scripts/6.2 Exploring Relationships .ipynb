#### 2. Exploring relationships

```python
# creating subset
df_UFC.columns
```

```python
df_sub = df_UFC[['b_wins', 'b_losses', 'b_height_cms',
       'b_Reach_cms', 'b_weight_lbs', 'r_wins', 'r_losses', 'r_height_cms',
       'r_Reach_cms', 'r_weight_lbs', 'b_age', 'r_age']]
```

```python
df_sub.corr()
```

```python
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

## Categorical Plot

```python
sns.histplot(df_UFC['r_Reach_cms'],bins = 20, kde = True)
```

```python
#Apply flags
df_UFC.loc[df_UFC['r_Reach_cms'] < 170, 'r_fighter_reach'] = 'Low reach'
```

```python
df_UFC.loc[(df_UFC['r_Reach_cms'] >= 170) & (df_UFC['r_Reach_cms'] < 200), 'r_fighter_reach'] = 'Middle reach'
```

```python
df_UFC.loc[df_UFC['r_Reach_cms'] >= 200, 'r_fighter_reach'] = 'High reach'
```

```python
df_UFC['r_fighter_reach'].value_counts(dropna = False)
```

```python
df_UFC.columns
```

```python
# Create a categorical plot in seaborn using the reach categories created above

sns.set(style="ticks")
g = sns.catplot(x="r_age", y="r_Reach_cms", hue="r_fighter_reach", data=df_UFC)
```

```python
# Create a categorical plot in seaborn using the reach categories created above
sns.set(style="ticks")
g = sns.catplot(x="r_win_by_decision_unanimous", y="r_Reach_cms", hue="r_fighter_reach", data=df_UFC)
```

Low reach category has less wins on redside where as middle reach has more wins which is possible due to the abundance of middle reach fighters. High Reach fighters on red side have just as many unanimous wins as low reach fighters.
